```python
# TODO: use pandasql to find seasoned reviewers
# Citation: https://www.techonthenet.com/sqlite/functions/julianday.php
time_query = """
select critic_name, cast(julianday(max(review_date))-julianday(min(review_date)) as int) date_diff
from reviews_df
where cast(strftime('%Y', review_date) as int)>= 2000 AND critic_name != 'Anonymous' 
group by critic_name
having date_diff>=3652
order by date_diff desc
"""
critics_time_df =  ps.sqldf(time_query, locals())
critics_time_df 
```

**TODO:** First, use **stopwords** to create a set of the most common english stopwords. Then, implement **tokenized_content(content)** that takes in a content string and:
1. tokenizes the text
2. lowercases the token
3. removes stop words (commonly used words such as "a","an", "in")
4. keeps words with only alphabet characters (no punctuation)

```python
import nltk
nltk.__version__
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
# Note that stopwords are all in lowercase format
```

```python
def tokenized_content(content):
  tokenized_sentence=[]
  for sentence in content:
    tokenized_sentence.append(nltk.word_tokenize(sentence))
  
  flat_list = []
  for element in tokenized_sentence:
    if type(element) is list:
     # If the element is of type list, iterate through the sublist
      
      for item in element:
        item=item.lower()
        if item not in stopwords and (item.isalpha()) == True:
           flat_list.append(item)
    else:
      element=element.lower()
      if element not in stopwords and (element.isalpha()) == True:
          flat_list.append(element.lower())
  
  return flat_list
```
